```python
from traced_v2.hierarchy import HashHierarchy
```

```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
h = HashHierarchy()
```

```python
words = [
    "h",
    "ha",
    "hat",
    "hello",
    "hate",
    "hell"   
]
```

```python
for i, word in enumerate(words):
    hash = h.hash(list(word))
    print(f"iteration {i}")
    print(word, hash)
    fig = plt.figure(figsize=(10, 10))
    nx.draw(h.hash_graph, with_labels=True)
    print()
```
