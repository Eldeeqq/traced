```python
import pandas as pd
```

```python
df = pd.read_csv('../data/num_of_records.csv',thousands=',')
```

```python
df['timestamp per day']=pd.to_datetime(df['timestamp per day'])
```

```python
df['Count of records'] = df['Count of records'].astype(int)
```

```python
df.columns = ['Date', 'Count of records']
```

```python
ax = df.plot(x='Date', y='Count of records', figsize=(15, 5), grid=True, legend=False, color='green')
ax.set_title('Number of records')
ax.set_ylabel('Number of records')
ax.set_xlabel('Date')
ax.figure.savefig('../results/figures/num_of_records.svg')
```

```python
df = pd.read_csv('../data/site_heatmap.csv',thousands=',')
```

```python
df.columns = ['destination', 'source', 'count']
```

```python
df['count'] = df['count'].astype(int)
```

```python
heatmap_data = df.pivot(index='destination', columns='source', values='count')
```

```python


import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
heatmap_data.sort_index(level=0, ascending=True, inplace=True)
```

```python

# Create a heatmap using Seaborn
fig = plt.figure(figsize=(8, 6))
sns.heatmap(heatmap_data, annot=False, cmap='flare', square=False)  # 'd' format for integer values
plt.xlabel('Destination')
plt.ylabel('Source')
plt.title('Heatmap')
plt.show()
fig.savefig('../results/figures/heatmap.svg')
```
